### task 1

```python
type1_instruct_fn = "../instruction_data/proc_instructions_type-i.csv"
type2_instruct_fn = "../instruction_data/proc_instructions_type-ii.csv"

type1_df = pd.read_csv(type1_instruct_fn, index_col=0)
type2_df = pd.read_csv(type2_instruct_fn, index_col=0)
tool_use_labels = [0] * len(type1_df) + [1] * len(type2_df)
```

```python
comb_df = pd.concat([type1_df[["did", "tool_name", "api_name", "instruction", "type"]], type2_df[["did", "tool_name", "api_name", "instruction", "type"]]], axis=0)
comb_df.reset_index(inplace=True, drop=True)
comb_df["tool_use_label"] = tool_use_labels
```

```python
comb_df.to_csv("output/eval_tasks/task1.csv")
```
