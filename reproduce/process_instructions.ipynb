```python
data_dir = os.path.join(os.getcwd(), "../instruction_data/") 

fn = os.path.join(data_dir, "dependency.csv") 
df = pd.read_csv(fn)
df = df[["source", "tool_name", "api_name", "dependencies", "comment"]]

sel_api_list = []
for idx in df.index:
    row = df.loc[idx]
    src, tool_name, api_name, dep, comment = row.tolist()
    if pd.isna(src):  # reaching the end of table
        break
    if pd.isna(dep) and pd.isna(comment):
        sel_api_list.append((src, tool_name, api_name))
```

```python
fn = os.path.join(data_dir, "dependency.csv")
df = pd.read_csv(fn)
df = df[["source", "tool_name", "api_name", "dependencies", "comment"]]

dep_api_list = []
for idx in df.index:
    row = df.loc[idx]
    src, tool_name, api_name, dep, comment = row.tolist()
    if pd.isna(src):  # reaching the end of table
        break
    if not pd.isna(dep):
        dep_api_list.append((src, tool_name, api_name, dep))
```
